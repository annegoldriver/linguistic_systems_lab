```python
import time
from datetime import datetime

# Setup Header
now = datetime.now()
timestamp = now.strftime('%m/%d/%Y')
header = 'Education Cultural Logic'
footer = 'Warning: Infinite Loop Detected in Agency Logic'

print(f'Date: {timestamp}')
print('\n' + '***' * len(header))
print(f'{header}')
print('***' * len(header))

# Education Cultural Logic - Feb 2026
def academic_journey_comparison():
    print('\n' + '--- The Formal college education "For Loop" System ---')
    # A For Loop: We know the steps (4 years/credits) and we iterate through them.
    years_in_college = ["Freshman", "Sophomore", "Junior", "Senior"]

    for year in years_in_college:
        print(f"Year: {year} - Completing credits, moving to the next stage.")
    print("Result: Degree Awarded (Iteration Complete).\n")

    print("--- The Vocational education 'While Loop' System ---")
    # A While Loop: Runs as long as a condition (Mastery/Duty) is NOT yet met.
    mastery_level = 0
    required_mastery = 100
    years_passed = 0

    # We keep going "while" we haven't reached the PhD level of mastery
    while mastery_level < required_mastery:
        years_passed += 1
        mastery_level += 25  # Gaining deep expertise each cycle
        print(f"Year {years_passed}: Practicing, researching, enduring (Mastery: {mastery_level}%)")
        time.sleep(0.5)

    print('Result: Mastery Achieved (Condition Finally Met).')


academic_journey_comparison()

print('\n' + '***' * len(header))
print(f'{footer}')
print('***' * len(header))


# Function Call: Call it once
academic_journey_comparison()

print('\n' + '--- INTERMISSION ---')

# Function Call: Call it again!
academic_journey_comparison()

```
