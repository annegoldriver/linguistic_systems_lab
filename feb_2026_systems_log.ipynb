```python
from datetime import datetime

now = datetime.now()
timestamp = now.strftime("%m/%d/%Y")

header = 'New OS installed, OS Vertias version 2026'
footer = 'Installed successfully'

environment = ['Asset','Virus', 'Asset', 'Error', 'Asset']
portfolio = []

for item in environment:
  if item != 'Virus' and item !='Error':
    portfolio.append(item)

print(timestamp)
print(header)
print(portfolio)
print("-" * 5, footer)

```

```python
from datetime import datetime

now = datetime.now()
timestamp = now.strftime('%m/%d/%Y')

header = 'New OS Initiative'
footer = 'Battery charged'

print(timestamp)
print('-' * len(header))
print(header)
print('-' * len(header))

missions = ['Definition', 'Decision', 'Freedom', 'Independence', 'Sovereignty']

for mission in missions:
    # Check if the current mission is one of the early steps
    if mission in ['Definition', 'Decision']:
        print(f'{mission}: Path to Freedom')

    # Check if we have reached the final goal
    elif mission == 'Sovereignty':
        print('STATUS: Sovereign OS Active')

print('-' * len(footer))
print(footer)
print('-' * len(footer))
```
