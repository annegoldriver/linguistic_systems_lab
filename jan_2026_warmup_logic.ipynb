**DateTime Syntax**
1. %Y	Full Year	2026	Year (Grandioso / 4-digits)
2. %y	young Year	26	young (Truncated / 2-digits)
3. %B	Full Month	January	Big name (Legato)
4. %b	brief Month	Jan	brief (Staccato)
5. %m	month number	01	month index (The "Scale" degree)
6. %A	All Weekday	Tuesday	All of the name
7. %a	abbreviated	Tue	abbreviated (Short)
8. %d	day number	06	date number (The "Beat")
9. %H	Hour (24)	14	Hour (High count / Military)
10. %I	Hour (12)	02	Index (Clock face / 1 to 12)
11. %p	AM/PM	AM	period of day
12. %M	Minute	45	Minute
13. %S	Second	30	Second


```python
# Daily Etude (practice script): Syntax Accuracy Drill

# The "Accuracy Etude"
from datetime import datetime

# 1. Careful with the symbols!
now_obj = datetime.now()
current_time = now_obj.strftime("%H:%M:%S")

# 2. Careful with the "F-string" and braces
header = f"Session started at: [{current_time}]"
print("*" * len(header)) # Line exactly as long as text !
print(header)
print("*" * len(header))

# 3. Careful with the indentation (The "Sub-division")
practice_set = ["Dynbamics", "Rhythm", "Articulation"]
for skill in practice_set:
  print(f"Practicing: {skill}...")
```

**1. Comment: #, """ """**
* #sinlge line comment
* """ triple quotes """: multiple lines comment

**2. f string modifier: Literal PREFIX: like -ed morpheme**
* with f string: live performance. The score reacts to the soloitst
* without f string: fixed recording. No interaction possible.

**3. f string logic of syntax: f + {}**
* f: formatted string
* " " double quotes : the boundaries of the string.
* {} braces: portals. inside them all as code, not just text.

```python
# single comment
"""
Multiple comment :
Triple quote
"""

# f string (formatted string): Prefix - a special modifier
"""
- with f text: live performance
- without f text: fixed recording
"""

name = "Maestro"
bpm = 120

# 1. No 'f' - The "Muted" String
print("Hello {name}, the tempo is {bpm}")

# 2. With 'f' - The "Live" String
print(f"Hello {name}, the tempo is {bpm}")

# 3. You can even do MATH inside the braces!
print(f"If we double the tempo, it's {bpm * 2} BPM.")
```
